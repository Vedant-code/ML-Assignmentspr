```python
plt.figure(figsize=(12, 6))
sns.barplot(x = data['SepalLengthCm'], y = data['PetalLengthCm'])
plt.title('Variation of Sepal Length with Petal Length', size=20)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = data['SepalLengthCm'])
plt.title('Appearance of values of Sepal Length', size = 20)
plt.xlabel('Value of Sepal Length', size=10)
plt.ylabel('Number of times the value appears', size=10)
plt.show()
```

```python
plt.boxplot(data['SepalLengthCm'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.title('Heat Map of the dataset', size=20)
plt.show()
```
