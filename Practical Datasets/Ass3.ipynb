```python
features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP',
       'LOR ', 'CGPA', 'Research']
x = data[features]

data['Chance of Admit '] = np.where(data['Chance of Admit ']>=0.9,1,0)
print(data['Chance of Admit '].head())
y = data['Chance of Admit ']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
model = DecisionTreeClassifier()
model = model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_train_pred = model.predict(x_train)
```

```python
#Training Accuracy
accuracy1 = metrics.accuracy_score(y_train, y_train_pred)
print(f'Training Accuracy = {accuracy1}')

#Testing Accuracy
accuracy2 = metrics.accuracy_score(y_test, y_pred)
print(f'Testing Accuracy = {accuracy2}')
```
